# 06_04 - Decision Engine: MDP Mathematical Formulation

## 1. Scope

This notebook translates the mathematical formulation from the PDF *Mathematical Formulation MDP* into executable project documentation.

The model defines a Markov Decision Process (MDP) for coordinated production and energy-procurement decisions under inventory, capacity, and market constraints.

## 2. System and Economic Parameters

The physical and economic constants used in the MDP are:

- Physical and financial limits: $P_{\max}=1500$, $D=1000$, $I_{\max}=3000$
- Energy coefficients: $e_{\text{start}}=20\,\text{MWh}$, $e_{\text{unit}}=1\,\text{MWh/unit}$
- Cost and margin terms: $h=5\,\text{EUR/unit/day}$, $M=200\,\text{EUR/unit}$

Interpretation:

- $P_{\max}$ is maximum daily production.
- $D$ is daily demand.
- $I_{\max}$ is maximum inventory.
- $h$ is holding cost per stored unit and day.
- $M$ is the unit contribution margin.

## 3. MDP Formulation

### 3.1 State Space $S$

At day $t$, the state vector is:

$$
s_t = [I_t, d_t, V_{M0,t}, V_{M1,t}, V_{M2,t}, V_{M3,t}, p_{DA,t+1}, p_{M1,t}, p_{M2,t}, p_{M3,t}]
$$

Where:

- $I_t$: inventory level
- $d_t$: day of month
- $V_{Mk,t}$: energy pool for month bucket $k$ (current month $M0$ and forwards $M1$ to $M3$)
- $p_{DA,t+1}$: day-ahead spot price for tomorrow
- $p_{Mk,t}$: forward prices for $M1$ to $M3$

### 3.2 Action Space $A$

The action vector is:

$$
a_t = [P_{t+1}, b_{M1,t}, b_{M2,t}, b_{M3,t}]
$$

With constraints:

- Trading block restriction:
$$
b_{Mk,t} \in \{0, 500, 1000\}, \quad k \in \{1,2,3\}
$$

- Production feasibility:
$$
\max(0, 1000 - I_t) \le P_{t+1} \le \min(1500, 3000 + 1000 - I_t)
$$

### 3.3 Cost and Reward Dynamics

Daily reward is defined as revenue minus storage cost, spot-energy cost, and futures expenditure:

$$
R_t = \text{Revenue} - \text{Storage Cost} - \text{Spot Energy Cost} - \text{Futures Expenditure}
$$

Equivalent expanded expression:

$$
R_t = (M \cdot P_{t+1}) - (h \cdot I_{t+1}) - (p_{DA,t+1} \cdot E_{\text{spot used}}) - \sum_{k=1}^{3} (p_{Mk,t} \cdot b_{Mk,t})
$$

This objective enforces the central trade-off of the policy: satisfy production and demand while minimizing procurement and inventory carrying costs under uncertain prices.
