## 1. Methodology Overview

This notebook regenerates the raw holidays file used by the project pipeline.

**Source implementation:**
- `src/data/generate_holidays_raw.py`

**Output generated:**
- `data/raw/holidays/holidays_raw.csv`

The holiday calendar is built with the `holidays` Python package for Spain (`country_code='ES'`) and constrained to years 2020 through 2025.

```python
from pathlib import Path
import sys

project_root = Path.cwd().resolve()
if not (project_root / 'src').exists():
    project_root = Path('../../').resolve()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.data.generate_holidays_raw import generate_holidays_raw
from src.config.paths import HOLIDAYS_RAW_FILE

holidays_df = generate_holidays_raw(start_year=2020, end_year=2025, country_code='ES')

print(f'Generated file: {HOLIDAYS_RAW_FILE}')
print(f'Rows: {len(holidays_df)}')
print(f'Date range: {holidays_df["date"].min().date()} -> {holidays_df["date"].max().date()}')
```
