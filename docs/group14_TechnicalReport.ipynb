    2.3 Final table

Once we have this information, the aggregated table has been done. We have decided not to include every province in the final dataframe, as we consider there are higly correlated.
So, to get every column of the final table, this considerations have been made:
- **Weighting average by Population**: Columns weather_code, temperature_2m_mean, temperature_2m_max, temperature_2m_min, apparent_temperature_mean*
- **Weighting average by Surface**: Columns *wind_speed_10m_max, wind_gusts_10m_max, shortwave_radiation_sum, precipitation_sum, rain_sum, snowfall_sum, precipitation_hours, surface_pressure_mean, et0_fao_evapotranspiration_sum*
- **Take Madrid (most important city and in the middle of the country)**: Columns *sunrise, sunset*
- **Not taken columns**: *daylight_duration* As it has a high correlation with sunrise and sunset columns

<br />

Also, some derived weather features have been added:
- **daily_temperature_range**: Maximum daily temperature - Minimum Daily temperature
- **rolling_avg_temp**: AVG temperature of last 3 days (not counting current).
- **delta_temp_with previous**: abs(Avg daily temp - rolling_avg_temp)

- **peninsular_max_temperature**: Maximum of all temperatures (excludind Canary Islands, Balearic Islands, Ceuta and Melilla).
- **peninsular_min_temperature**: Minimum of all temperatures (excludind Canary Islands, Balearic Islands, Ceuta and Melilla)
- **max_windspeed**: As the eolic energy does not grow linearly (in a cubic way), the maximum speed can contribute quite a lot 

**When taking maximums and minimums, the 3th biggest number has been taken (not the first and second to mitigate outliers)**

<br />

Also, following columns have been added in order to give more valuable information
- **Day_of_the_week**: (1 for Monday, 7 for Sunday)
- **Is_weekend**: (Binary variable)
