```python
import urllib.request
import urllib.parse
import json

# Setting up the weather data input parameters
api_key = '9N875PXLR8DRBEALPHANESGEL'
location = 'maçanet de la selva'

# The specific date you want to travel back in time to
basis_date = '2024-01-01'

# We ask for that day, plus the following days to make a week
start_date = '2024-01-01'
end_date = '2024-01-07'

unit_group = 'metric'
content_type = 'json'
include = 'days'

# URL-encode the location to handle spaces and special characters
encoded_location = urllib.parse.quote(location)

# Constructing the URL for the API request
base_url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{encoded_location}/{start_date}/{end_date}'

# THIS IS THE SECRET SAUCE: Adding forecastBasisDate to the query parameters
query_parameters = f'?unitGroup={unit_group}&include={include}&key={api_key}&contentType={content_type}&forecastBasisDate={basis_date}'

url = base_url + query_parameters
print(f"Constructed URL: {url}")  # Notice the forecastBasisDate at the end!

try:
    # Requesting data from the API
    response = urllib.request.urlopen(url)
    data = response.read().decode('utf-8')
    
    # Parsing JSON data
    weather_data = json.loads(data)
    
    # Example of accessing data
    print(json.dumps(weather_data, indent=4))
    
except urllib.error.HTTPError as e:
    # Handling HTTP errors
    print(f'HTTP Error: {e.code} - {e.reason}')
    print(e.read().decode('utf-8'))  
except urllib.error.URLError as e:
    # Handling URL errors
    print(f'URL Error: {e.reason}')
```
